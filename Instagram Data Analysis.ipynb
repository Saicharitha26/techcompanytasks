```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['Likes'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Likes on Instagram Posts')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Likes', y='Comments', data=df)
plt.title('Likes vs Comments')
plt.xlabel('Likes')
plt.ylabel('Comments')
plt.tight_layout()
plt.show()
```
